```
import pandas as pd


# Criando as listas(Colunas)
Modelo =["Civic SI",'Silvia - S15','Rx7','GT-86','AMG-GT','RS-6','M3- Competition','Golf']
Marca =['Honda','Nissan','Mazda','Toyota','Mercedes','Audi','BMW','Volkswagem']
potencia=['208 cv','197 cv','280 cv','199 cv','816 cv','630 cv','510 cv','242 cv']
ano_lancamento=[2007, 1999, 1978, 2013, 2014, 2009, 2021,1974]
preco=['R$72.900','R$87.894','R$48.862','R$134.490','R$2.434.246','R$1.175.000','R$829.950','R$590.000']
carroceria=['SEDAN','SEDAN','SEDAN','PERUA','COUPE','PERUA','SEDAN','HATCHBACK']


# Criando uma biblioteca com as listas
carros={
    'Modelo':Modelo,
    'Marca' :Marca,
    'Potencia':potencia,
    'Ano de lançamento':ano_lancamento,
    'Preço':preco,
    "Carroceria":carroceria
}


# Criando o DF  e seus indexs a partir da biblioteca carros
museu_carros = pd.DataFrame(carros, index=["Carro 1",'Carro 2','Carro3','Carro 4','Carro 5' ,'Carro 6','Carro 7','Carro 8'])


#Mostra o DF
display(museu_carros)
print()
```

```
# Mostra o tamanho do DF (linhas e colunas)
print('\n\n1) Apresente o tamanho do seu dataframe (quantas colunas x linhas).\n')
print(museu_carros.shape)
print()
```

```
# Acessando a linha "Carro 1" e mostrando seus atributos
print('\n\n2) Acesse a linha (x) e apresente em tela todas as características do item.\n')
linha_carro1 = museu_carros.loc["Carro 1"]
display(linha_carro1)
print()
```

```
#verificando se os DF esta vazio ou não
print('\n\n 3) Verifique se o dataframe está vazio.\n')
if museu_carros.empty:
    print("O DataFrame está vazio.\n\n")
else:
    print("O DataFrame não está vazio.\n\n")
```

```
#Apresentando os 5 primeiros registros
print('\n\n4) Apresente em tela os 5 primeiros registros da base de dados.\n')
display(museu_carros.head())

print('\n\n')
```

```
print('\n\n5)Exclua um item (linha) de sua base de dados.')
museu_carros.drop("Carro 9", inplace=True)
display(museu_carros)
```

```
print('\n\n6)Adicione um item (linha) na sua base de dados\n')
novo_carro = {
  'Modelo': 'Fiesta',
  'Marca': 'Ford',
  'Potencia': '120 cv',
  'Ano de lançamento': 2022,
  'Preço': 'R$65.000',
   'Carroceria': 'HATCHBACK'
}


 #Adicionando uma nova linha ao DataFrame
museu_carros.loc['Carro 9'] = novo_carro


 #Exibindo o DataFrame após a adição
display(museu_carros)
```

```
# Exibindo a base de dados transposta
museu_carros_transposto = museu_carros.transpose()
print('\n\n7) Transponha a coluna para a linha em sua base de dados.\n')
display(museu_carros_transposto)
print('\n\n')
```

```
# Exibindo a primeira e a segunda coluna
print('\n\n8) Apresente em tela somente a 1ª e a 2ª coluna (rótulo) da base de dados.\n')
primeira_segunda_coluna = museu_carros.iloc[:, :2]
display(primeira_segunda_coluna)
```
